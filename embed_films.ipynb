```python
import pycountry
```

```python
def get_language_name(code):
    if pd.isna(code):
        return "Unknown"
    try:
        # Pobieramy jÄ™zyk na podstawie kodu alpha_2 (np. 'en')
        lang = pycountry.languages.get(alpha_2=code)
        return lang.name if lang else code
    except AttributeError:
        return code
    
df['original_language'] = df['original_language'].apply(get_language_name)

print(df['original_language'].head())
```

```python
# def create_text_chunk(row):
#     parts = [
#         f"Title: {row['title']}",
#         f"Original title: {row['original_title']}",
#         f"Genres: {row['genres']}",
#         f"Keywords: {row['keywords']}",
#         f"Tagline: {row['tagline']}",
#         f"Production: {row['production_companies']}",
#         f"Overview: {row['overview']}"
#     ]
#     return ". ".join([p for p in parts if p])

def create_text_chunk(row):
    # Lista czÄ™Å›ci skÅ‚adowych opisu
    parts = []
    
    # 1. TytuÅ‚ i Rok (Kontekst podstawowy)
    title = row['title']
    year = int(row['year']) if pd.notnull(row['year']) else ""
    if year:
        parts.append(f"Movie title: {title} ({year}).")
    else:
        parts.append(f"Movie title: {title}.")

    # 2. Oryginalny tytuÅ‚ (jeÅ›li inny)
    if pd.notnull(row['original_title']) and row['original_title'] != title:
        parts.append(f"Original title: {row['original_title']}.")

    # 3. JÄ™zyki (NOWOÅšÄ† - wymagane przez tabelÄ™)
    lang_info = []
    if pd.notnull(row['original_language']):
        lang_info.append(f"Original language: {row['original_language']}")
    if pd.notnull(row['spoken_languages']):
        lang_info.append(f"Spoken languages: {row['spoken_languages']}")
    
    if lang_info:
        parts.append(". ".join(lang_info) + ".")

    # 4. Tagline (Klimat)
    if pd.notnull(row['tagline']) and len(str(row['tagline'])) > 3:
        parts.append(f"Tagline: {row['tagline']}")

    # 5. Gatunki (Kategoryzacja)
    if pd.notnull(row['genres']):
        parts.append(f"Genres: {row['genres']}.")

    # 6. FabuÅ‚a (TreÅ›Ä‡ gÅ‚Ã³wna)
    if pd.notnull(row['overview']):
        parts.append(f"Plot summary: {row['overview']}")

    # 7. Keywords (SzczegÃ³Å‚y tematyczne)
    if pd.notnull(row['keywords']):
        parts.append(f"Keywords: {row['keywords']}.")

    # 8. Kontekst produkcji
    prod_info = []
    if pd.notnull(row['production_companies']):
        prod_info.append(f"Produced by: {row['production_companies']}")
    if pd.notnull(row['production_countries']):
        prod_info.append(f"Country: {row['production_countries']}")
    
    if prod_info:
        parts.append(" | ".join(prod_info) + ".")

    # ÅÄ…czymy w jeden spÃ³jny tekst
    return " ".join(parts)
```

```python
df["text_to_embed"] = df.apply(create_text_chunk, axis=1)
```

```python
from qdrant_client import QdrantClient
from qdrant_client.models import VectorParams, Distance, PointStruct, SparseVectorParams, SparseVector
from fastembed import SparseTextEmbedding
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import torch
```

```python
# model = SentenceTransformer('all-MiniLM-L6-v2', device='mps')
dense_model = SentenceTransformer('Qwen/Qwen3-Embedding-0.6B', trust_remote_code=True, device='mps')
sparse_model = SparseTextEmbedding(model_name="Qdrant/bm25")
```

```python
BATCH_SIZE = 24
total_rows = len(df)

try:
    collection_info = client.get_collection(COLLECTION_NAME)
    current_count = collection_info.points_count
    print(f"W bazie jest juÅ¼ {current_count} filmÃ³w. Wznawiam od tego momentu.")
except Exception:
    current_count = 0
    print("Kolekcja nie istnieje lub jest pusta. Zaczynam od 0.")
```

```python
def parse_list(val):
    if not val: return []
    return [x.strip() for x in str(val).split(',')]
```

```python
for start_idx in tqdm(range(current_count, total_rows, BATCH_SIZE)):
    end_idx = min(start_idx + BATCH_SIZE, total_rows)
    batch_df = df.iloc[start_idx:end_idx]
    
    texts = batch_df["text_to_embed"].tolist()
    dense_vectors = dense_model.encode(texts, show_progress_bar=False)
    sparse_vectors = list(sparse_model.embed(texts))
    
    points = []
    for i, (_, row) in enumerate(batch_df.iterrows()):
        
        payload = {
            # Identyfikacja i UI
            "id": int(row['id']),
            "title": row['title'],
            "original_title": row['original_title'],
            "overview": row['overview'], # Potrzebne, Å¼eby LLM miaÅ‚ z czego generowaÄ‡ odpowiedÅº!
            "tagline": row['tagline'],
            "poster_path": row['poster_path'],
            "backdrop_path": row['backdrop_path'],
            "imdb_id": row['imdb_id'],
            
            # Filtrowanie (Numeryczne/Boolean)
            "year": int(row['year']) if pd.notnull(row['year']) else None,
            "vote_average": float(row['vote_average']),
            "vote_count": int(row['vote_count']),
            "popularity": float(row['popularity']),
            "runtime": int(row['runtime']),
            "adult": bool(row['adult']),
            
            # Filtrowanie (Listy/Kategorie)
            # ZakÅ‚adam, Å¼e w CSV genres/keywords to stringi oddzielone przecinkami
            "genres": [g.strip() for g in str(row['genres']).split(',')] if pd.notnull(row['genres']) else [],
            "keywords": [k.strip() for k in str(row['keywords']).split(',')] if pd.notnull(row['keywords']) else [],
            "production_companies": [g.strip() for g in str(row['production_companies']).split(',')] if pd.notnull(row['production_companies']) else [],
            "production_countries": [g.strip() for g in str(row['production_countries']).split(',')] if pd.notnull(row['production_countries']) else [],
            "spoken_languages": [g.strip() for g in str(row['spoken_languages']).split(',')] if pd.notnull(row['spoken_languages']) else [],
            "original_language": str(row['original_language'])
        }
        
        sparse_vec_obj = sparse_vectors[i]
        qdrant_sparse_vector = SparseVector(
            indices=sparse_vec_obj.indices.tolist(),
            values=sparse_vec_obj.values.tolist()
        )

        points.append(PointStruct(
            id=int(row['id']),
            vector={
                "text-dense": dense_vectors[i].tolist(),
                "text-sparse": qdrant_sparse_vector
            },
            payload=payload
        ))
    client.upsert(
        collection_name=COLLECTION_NAME,
        points=points
    )
    del dense_vectors, sparse_vectors, points, texts
    if torch.mps.is_available:
        torch.mps.empty_cache()
```
